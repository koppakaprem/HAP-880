```python
from sklearn.metrics import accuracy_score
test_acc=[]
train_acc=[]
aucTest=[]
aucTrain=[]
Clist = []
#testing data
for c in param["C"]:
    logreg = LogisticRegression(C=c,penalty='l1')
    Clist.append(c)
    logreg.fit(trainsplit[attributes],trainsplit['HighUtilizationY2'])
    probs = logreg.predict_proba(testsplit[attributes])[:,1]
    fpr, tpr, thresholds = roc_curve(testsplit['HighUtilizationY2'], probs)
    a=auc(fpr,tpr)
    aucTest.append(a)
    prcls= logreg.predict(testsplit[attributes]) 
    accScore = accuracy_score(testsplit['HighUtilizationY2'], prcls)
    test_acc.append(accScore)
    #training data
    probs = logreg.predict_proba(trainsplit[attributes])[:,1]
    fpr, tpr, thresholds = roc_curve(trainsplit['HighUtilizationY2'], probs)
    a=auc(fpr,tpr)
    aucTrain.append(a)
    prcls = logreg.predict(trainsplit[attributes]) 
    accScore = accuracy_score(trainsplit['HighUtilizationY2'], prcls)
    train_acc.append(accScore)
```
